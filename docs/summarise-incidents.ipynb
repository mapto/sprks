```python
import json
import glob
from IPython.display import HTML
```

```python
path = "../static/incidents/*.json"
rowt = """
<td>{id}</td>
<td>{name}</td>
<td>{policy}</td>
<td>{description}</td>
<td>{event}</td>
<td>{consequences}</td>
<td>{fx}</td>
"""
tablet = "<table><tr style='font-weight: bolder;'>{heading}</tr>{data}</table>"
```

```python
data = ""
itemnames = {}
for f in glob.glob(path):
    j = json.load(open(f, "r"))

    # merge numeric items
    values = ["risk", "cost"]
    values += [f"{v}-modifier" for v in values]
    j["fx"] = {v: j[v] for v in values}
    for v in values:
        j.pop(v)

    # get item after cleanup
    if not itemnames:
        itemnames = dict(zip(j.keys(),j.keys()))

    data += f"<tr>{rowt.format(**j)}</tr>"
#print(itemnames)
HTML(tablet.format(heading=rowt.format(**itemnames), data=data))
```
